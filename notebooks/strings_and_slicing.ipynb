str instance doesn't have associated binary encoding; bytes instance doesn't have associated text encoding. In order to convert unicode -> binary you must call the decode method of bytes. two helper functions:

```python

def to_str(bytes_or_str): 
    
    if isinstance(bytes_or_str, bytes):
        value = bytes_or_str.decode("utf-8")
    else: 
        value = bytes_or_str
    return value # instance of string 


def to_bytes(bytes_or_str):
    
    if isinstance(bytes_or_str, str):
        value = bytes_or_str.encode("utf-8")
    else:
        value = bytes_or_str # instance of bytes 
        

```
